```python
import numpy as np

# Ejercicio 1
ej1 = np.zeros(10)
print("Ej1:", ej1)
```

```python
# Ejercicio 2
ej2 = np.arange(1, 21)
print("Ej2:", ej2)

```

```python
# Ejercicio 3
ej3 = np.arange(2, 51, 2)
print("Ej3:", ej3)
```

```python
# Ejercicio 4
ej4 = np.random.rand(4,4)
print("Ej4:\n", ej4)
```

```python
# Ejercicio 5
a = np.array([1,2,3])
b = np.array([4,5,6])
ej6 = a + b
print("Ej5:", ej5)
```

```python
# Ejercicio 6
m1 = np.random.randint(1, 11, (3,3))
m2 = np.random.randint(1, 11, (3,3))
ej7 = np.dot(m1, m2)
print("Matriz 1:\n", m1)
print("Matriz 2:\n", m2)
print("Ej6 (multiplicación):\n", ej6)
```

```python
# Ejercicio 7
ej8 = np.random.randint(1, 101, 15)
print("Ej7:", ej7)
print("Máximo:", np.max(ej8))
print("Mínimo:", np.min(ej8))
print("Promedio:", np.mean(ej8))
```

```python
# Ejercicio 8
ej9 = np.arange(1, 26).reshape(5,5)
diagonal = np.diag(ej9)
print("Ej8:\n", ej8)
print("Diagonal:", diagonal)
```

```python
# Ejercicio 9
mat = np.random.randint(1, 21, (5,5))
print("Matriz:\n", mat)
print("Suma filas:", np.sum(mat, axis=1))
print("Suma columnas:", np.sum(mat, axis=0))
```
